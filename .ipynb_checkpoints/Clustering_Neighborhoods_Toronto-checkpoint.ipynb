# Clustering Neighborhoods in Toronto 
This notebook shows how to scrape web data about boroughs and neighborhoods in Toronto, enrich it with location information from Foursquare and cluster the neighborhoods based on the types of venues in them. 

## Scraping data from Wikipedia and preparing a dataframe

```python
#Import packages
import numpy as np 
import pandas as pd
import requests 
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
import lxml
import json
#!conda install -c conda-forge geopy --yes # uncomment this line if necessary
from geopy.geocoders import Nominatim 
#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if necessary
from bs4 import BeautifulSoup
import folium
print('Libraries imported.')
```

### Data Scraping 
The data is scraped from the table on the following Wikipedia page: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M. 

The BeautifulSoup package is used to obtain the data that is in the table and to transform the data into a Pandas data-frame. 

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
df_scraped = pd.read_html(tab)
df=df_scraped[0]
df.head(40)


```

### Conversion of html tab to Pandas dataframe for data cleaning and preparation
Postalcodes where no borough is assigned are dropped from the data-set. Neighborhoods with the same postalcode are combined, names of neighborhoods which are "Not assigned" are replaced with the names of the boroughs. 

```python
df = df[df.Borough != 'Not assigned']

df = df.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)

df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned',df['Borough'], df['Neighborhood'])

display(df.head(20))
```

```python
df.shape
```

## Data preparation for clustering
I download a data set with the geographical coordinates of each postal code and match it with the existing data set. 

```python
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
display(coordinates.head())

df_toronto_full = pd.merge(df,coordinates,on='Postal Code')
display(df_toronto_full)
```

### Enrich the data-set with location information from the Foursquare data-base
Using the Foursquare API to get information about venues in the neighborhoods to generate a data-set for the clustering. Analysis is limited to those boroughts with Toronto in the name.

```python
# Define function to get top 50 nearby food venues (radius=500m) for all neighborhoods
def getNearbyVenues(names,
                    latitudes,
                    longitudes,
                    radius=500,
                    LIMIT=50,
                    section="food"):

    venues_list = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)

        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&section={}'.format(
            CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT, section)

        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']

        # return only relevant information for each nearby venue
        venues_list.append([
            (name, lat, lng, v['venue']['name'], v['venue']['location']['lat'],
             v['venue']['location']['lng'],
             v['venue']['categories'][0]['name']) for v in results
        ])

    nearby_venues = pd.DataFrame(
        [item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = [
        'Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude',
        'Venue', 'Venue Latitude', 'Venue Longitude', 'Venue Category'
    ]

    return (nearby_venues)
```

```python
#Applying the function to get the top 50 food venues for all neighborhoods
toronto_venues = getNearbyVenues(names=df_toronto['Neighborhood'],
                                   latitudes=df_toronto['Latitude'],
                                   longitudes=df_toronto['Longitude']
                                  )
```

```python
# Check resulting dataframe
display(toronto_venues.shape)
display(toronto_venues.head())
pd.set_option("display.max_rows", 100, "display.max_columns", None)
display(toronto_venues.groupby('Neighborhood').count())
```

### Prepare the data-set for cluster analysis
Transform the venue category variable into a set of dummy variables and calculate the mean frequency of different venue types for further analysis. 

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']],
                                prefix="",
                                prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(
    toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

display(toronto_onehot.head())
display(toronto_onehot.shape)
```

```python
#Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category (mean of dummies is the percentage this category)
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
#print each neighborhood along with the top 5 most common venues
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','perc']
    temp = temp.iloc[1:]
    temp['freq'] = temp['perc'].astype(float)
    temp = temp.round({'perc': 2})
    print(temp.sort_values('perc', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Clustering neighborhoods in Toronto 
Cluster analysis based on percentage of specific restaurant types among the top 5 restaurants in the neighborhoods. 

```python
# Data-set is toronto_grouped, k-means clustering

# Clusteranalysis
# set number of clusters
kclusters = 11

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 


```

```python
# add clustering labels to toronto_grouped (dataset with neighborhood and venue categories as dummy variables)
toronto_clusters = toronto_grouped
toronto_clusters.insert(0, 'Cluster Labels', kmeans.labels_)
display(toronto_clusters.head())


```

```python
# merge toronto_clusters with location data from toronto_data to add latitude/longitude for each neighborhood
toronto_geoloc = df_toronto.drop(columns=['Postal Code', 'Borough'])
toronto_clusters = toronto_clusters.join(toronto_geoloc.set_index('Neighborhood'), on='Neighborhood')

toronto_clusters.head() # check the last columns!
```

### Examine clusters
Take a closer look at the clusters by mapping the resulting clusters and by looking at the top 5 most common venues in each cluster.

#### Mapping clusters

```python
# create map
map_clusters = folium.Map(location=[43.8184944, -79.3309468], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_clusters['Latitude'], toronto_clusters['Longitude'], toronto_clusters['Neighborhood'], toronto_clusters['Cluster Labels']):
    label = folium.Popup(str(poi) + ' | Cluster: ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
#Group rows by cluster number and by taking the mean of the frequency of occurrence of each category
toronto_clusterexam = toronto_clusters.groupby('Cluster Labels').mean().reset_index().drop(columns=['Latitude', 'Longitude'])
toronto_clusterexam['Cluster Labels'] = toronto_clusterexam['Cluster Labels'].astype(str)
toronto_clusterexam
```

```python
#print each cluster along with the top 5 most common venues

num_top_venues = 5

for cluster in toronto_clusterexam['Cluster Labels']:
    print("----"+cluster+"----")
    temp = toronto_clusterexam[toronto_clusterexam['Cluster Labels'] == cluster].T.reset_index()
    temp.columns = ['venue','perc']
    temp = temp.iloc[1:]
    temp['perc'] = temp['perc'].astype(float)
    temp = temp.round({'perc': 2})
    print(temp.sort_values('perc', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```
